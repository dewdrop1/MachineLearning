### Generate some artifical data

In supervised modeling, the data has inputs and output labels.

A model is a function that maps an input x to an output y. The model's output is called a prediction, denoted by y_hat.

A model is evaluated by a measurement called a loss or error. MSE (mean squared error) is an example metric.

A model is as good as its ability to genaralize to new examples.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_artificial_data(num_examples=100, m=1, b=2):
   X = np.arange(num_examples)
   np.random.seed(4)
   deltas = np.random.uniform(low=-1,high=1,size=X.shape)
   Y = b + deltas + m * X
   return X, Y
```

```python
X, Y = create_artificial_data()
plt.scatter(X, Y)
plt.show()
```

```python
# model1
# M(x) = x + 5

def model_1(X):
  y_hat = X + 5
  return y_hat

def model_2(X):
  y_hat = 2*X + 1
  return y_hat

def compute_mse(y, y_hat):
  return 1/y.size * ((y - y_hat)**2).sum()
```

```python
y_model_1 = model_1(X)
```

```python
y_model_2 = model_2(X)
```

```python
plt.scatter(X, Y, c = 'blue', label='actual')
plt.plot(y_model_1, 'C1', label='model 1')
plt.plot(y_model_2, 'C2', label = 'model 2')
plt.legend()
plt.show()
```

```python
compute_mse(Y, model_1(X))
```

```python
compute_mse(Y, model_2(X))
```
